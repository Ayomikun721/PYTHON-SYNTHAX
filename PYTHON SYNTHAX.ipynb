```python
import pandas as pd
import random
import matplotlib.pyplot as plt
```

```python
SalesPerson = []
Person = str(input('Sales Person: '))
SalesPerson.append(Person)

ItemName = []
Price = []
Profit = []
SalesPrice = []
QtySupplied = []
Supplier = []
InvoiceNo = []
ExpIncome = []
ExpProfit = []

StockList = []
UpdateList1 = []
TotalqtySupp = []
Totalprofitexp = []
Totalincomeexp= []

def Stock():
    no = int(input('Number of Items: '))
    for i in range(no):
        n=input('Item: ').upper()
        y=int(input('Quantity: '))
        if n in ItemName:
            index=ItemName.index(n)
            QtySupplied[index] += y
            qty=QtySupplied[index]
            pr=Profit[index]
            NewPrice=SalesPrice[index]
            totalexpProfit=round(float(qty)*pr,2)
            expIncome=round(float(NewPrice)*float(qty),2)
            ExpProfit[index]=totalexpProfit
            ExpIncome[index]=expIncome

            for i in StockList:
                if i['Item']==n:
                    i['Quantity Supplied']=qty
            for i in UpdateList1:
                if i['Item']==n:
                    i['Quantity Supplied']=qty
                    i['Total Profit Expected (NGN)']=totalexpProfit
                    i['Expected Income']=expIncome

        else:
            name=n
            qty=y
            pri=float(input('Price: NGN '))
            pr=0.2*pri
            NewPrice=pri+pr
            supplier=input('Supplier: ')
            InvNo=int(input('Invoice Number: '))
            
            ItemName.append(name)
            QtySupplied.append(qty)
            Price.append(pri)
            Profit.append(pr)
            SalesPrice.append(NewPrice)
            Supplier.append(supplier)
            InvoiceNo.append(InvNo)
            totalexpProfit=round(float(qty)*pr,2)
            expIncome=round(float(NewPrice)*float(qty),2)
            ExpProfit.append(totalexpProfit)
            ExpIncome.append(expIncome)
            
            Stock_dict={
                'Item': name,
                'Supplier': supplier,
                'Invoice Number': InvNo,
                'Quantity Supplied': qty,
                'Original Price Per Unit':pri,
                'Profit Per Unit': pr,
                'Sales Price Per Unit (NGN)': NewPrice
                
            }
            StockList.append(Stock_dict)
            UpdateStockDict={
                'Item': name,
                'Supplier': supplier,
                'Invoice Number': InvNo,
                'Quantity Supplied': qty,
                'Sales Price(NGN)': NewPrice,
                'Total Profit Expected': totalexpProfit,
                'Expected Income': expIncome
            }
            UpdateList1.append(UpdateStockDict)
    totalqtySupp=sum(QtySupplied)
    tpe=sum(ExpProfit)
    totalprofitexp=round(tpe,2)
    tie=sum(ExpIncome)
    totalincomeexp=round(tie,2)
    TotalqtySupp.append(totalqtySupp)
    Totalprofitexp.append(totalprofitexp)
    Totalincomeexp.append(totalincomeexp)

def df1():
    df1=pd.DataFrame(UpdateList1)
    return df1

def StockRecord():
    StockRecord=pd.DataFrame(StockList)
    return StockRecord
```

```python
RecordProduct=[]
RecordQuantity=[]
InvoiceOfProduct=[]
PriceRecord=[]
RecordProfit=[]
RecordAmount=[]
SalesList=[]
TotQtySold=[]
TotProfitMade=[]

def Sales_Record(SaleProduct, SaleQty):
    print(SalesPerson)
    c={}
    for d,n in zip(SaleProduct, SaleQty):
        if d not in c:
            c[d]=n
        else:
            c[d] += n
    for product, quantity in c.items():
        RecordProduct.append(product)
        RecordQuantity.append(quantity)
    for k in RecordProduct:
        index1=ItemName.index(k)
        IOP=InvoiceNo[index1]
        PR=SalesPrice[index1]
        profit_index=Profit[index1]
        index2=RecordProduct.index(k)
        qty=float(RecordQuantity[index2])
        RProfit=profit_index*qty
        RA=qty*PR

        InvoiceOfProduct.append(IOP)
        PriceRecord.append(PR)
        RecordProfit.append(RProfit)
        RecordAmount.append(RA)

        Sales_dict={
            'Product':k,
            'Invoice Number': IOP,
            'Price':PR,
            'Quantity': qty,
            'Amount': RA,
            'Profit Actualized': RProfit
        }
        SalesList.append(Sales_dict)

    TotalProfitMade=round(sum(RecordProfit),2)
    TotalQuantitySold=sum(RecordQuantity)
    TotQtySold.append(TotalQuantitySold)
    TotProfitMade.append(TotalProfitMade)
    print('Total Quantity Sold = {}'.format(TotalQuantitySold))
    print(f'{'Total Profit Made = '}{TotalProfitMade}')

def DataFrame():
    SalesRecord=pd.DataFrame(SalesList)
    return(SalesRecord)
```

```python
UpdateList2 = []
QuantitySold = []
ProfitAct = []
IncomeAct = []
QtyRem = []
ProfitRem = []
IncomeRem = []

for i in ItemName:
    if i in RecordProduct:
        index=RecordProduct.index(i)
        quantitySold=RecordQuantity[index]
        pA=RecordProfit[index]
        iA=RecordAmount[index]
    else:
        quantitySold=float(0)
        pA=float(0)
        iA=float(0)
    profitAct=round(pA,2)
    incomeAct=round(iA,2)
    QuantitySold.append(quantitySold)
    ProfitAct.append(profitAct)
    IncomeAct.append(incomeAct)
    
    index1=ItemName.index(i)
    OGQty=float(QtySupplied[index1])
    qtyRem=OGQty-quantitySold
    OGExpProfit=ExpProfit[index1]
    pR=OGExpProfit-pA
    profitRem=round(pR,2)
    OGExpectedIncome=ExpIncome[index1]
    iR=OGExpectedIncome-iA
    incomeRem=round(iR,2)
    QtyRem.append(qtyRem)
    ProfitRem.append(profitRem)
    IncomeRem.append(incomeRem)

    UpdateDict2={
        'QuantitySold':quantitySold,
        'Profit Actualized':profitAct,
        'Income Actualized':incomeAct,
        'Quantity Remaining':qtyRem,
        'Profit Remaining':profitRem,
        'Income Remaining':incomeRem
    }
    UpdateList2.append(UpdateDict2)

TotIncomeMade=round(sum(IncomeAct),2)
TotQtyRem=sum(QtyRem)
TotProfitRem=round(sum(ProfitRem),2)
TotIncomeRem=round(sum(IncomeRem),2)

def df2():
    df2=pd.DataFrame(UpdateList2)
    return df2

def Update_DataFrame():
    UpdateRecord=pd.concat([df1(),df2()],axis=1)
    date=input('Enter date:')
    UpdateRecord.to_excel('Project/{}.xlsx'.format(date))
    return UpdateRecord
```

```python
def Update_Essay():
    print('Total Quantity Supplied = {}'.format(TotalqtySupp[-1]))
    print('Total Profit Expected = NGN {}'.format(Totalprofitexp[-1]))
    print('Total Income Expected = NGN {}'.format(Totalincomeexp[-1]))
    print('Total Quantity Sold = {}'.format(TotQtySold[-1]))
    print('Total Profit Actualized = NGN {}'.format(TotProfitMade[-1]))
    print('Total Income Actualized = NGN {}'.format(TotIncomeMade))
    print('Total Quantity Remaining = {}'.format(TotQtyRem))
    print('Total Profit Remaining = NGN {}'.format(TotProfitRem))
    print('Total Income Remaining = NGN {}'.format(TotIncomeRem))
```

```python
def Update_Visual():
    fig, ax = plt.subplots(1, 3, figsize=(12, 4))
    sold=TotQtySold[-1]
    ax[0].pie([sold, TotQtyRem], labels=["Sold", "Remaining"], colors=['#2E7D32', '#EF5350'], autopct="%1.1f%%", startangle=90)
    ax[0].set_title("Quantity")

    actualized=TotProfitMade[-1]
    ax[1].pie([actualized, TotProfitRem],labels=["Actualized", "Remaining"],colors=['#F9A825', '#EF5350'], autopct="%1.1f%%",startangle=90)
    ax[1].set_title("Profit")

    ax[2].pie([TotIncomeMade, TotIncomeRem],labels=["Actualized", "Remaining"],colors=['#1565C0', '#EF5350'], autopct="%1.1f%%",startangle=90)
    ax[2].set_title("Income")

    for a in ax:
        a.axis("equal")

    plt.tight_layout()
    plt.show()
```
